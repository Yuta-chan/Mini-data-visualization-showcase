# Contour Plot

Visualización de un Contour Plot mediante la librería [Matplotlib](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html) de python.

Los datos se han generado sintéticamente a partir de los valores Z de una función que depende de dos variables x,y.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def f(x, y):
    return np.sin(x) + np.cos(10 + y * x) * np.cos(x)
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar()
plt.title('Contour Plot de $\sin{x} + \cos{(10+y\cdot x)} \cdot \cos{x}$')
plt.savefig('contourplot.png')
```
