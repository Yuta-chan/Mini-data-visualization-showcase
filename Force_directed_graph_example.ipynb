# FORCE-DIRECTED GRAPH

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()
G.add_edge('Patri', 'B')
G.add_edge('B', 'C')
G.add_edge('C', 'Patri')
G.add_edge('Patri', 'D')

# Dibujar el grafo utilizando el algoritmo de Fruchterman-Reingold
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1000, edge_color='k', font_size=15)

# Guardar el grafo como PNG
plt.savefig("graph.png")

# O como SVG para mejor calidad en entornos web
plt.savefig("graph.svg")
```

```python
pos = nx.random_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=700, edge_color='k', linewidths=1, font_size=15)
```

```python
from d3graph import d3graph, vec2adjmat

# Create example network
source = ['node A','node F','node B','node B','node B','node A','node C','node Z']
target = ['node F','node B','node J','node F','node F','node M','node M','node A']
weight = [5.56, 0.5, 0.64, 0.23, 0.9, 3.28, 0.5, 0.45]
# Convert to adjacency matrix
adjmat = vec2adjmat(source, target, weight=weight)

# Initialize
d3 = d3graph()
# Proces adjmat
d3.graph(adjmat)
# Plot
d3.show()

# Make changes in node properties
d3.set_node_properties(color=adjmat.columns.values, label=['node 1','node 2','node 3','node 4','node 5','node 6','node 7'])
# Plot
d3.show()
```
